```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

# All Combined

```python
df = pd.read_excel('/content/All combined.xlsx',usecols=['date','tweet','event','sentiment'])
```

```python
df['sentiment'] = df.sentiment.map({'neutral':2, 'negative':0, 'positive':1}) #or{'neutral':0, 'negative':-1, 'positive':1}
df
```

```python
sentiment = df.groupby(['date', 'sentiment']).sentiment.count().unstack()
sentiment.plot(kind='area',figsize=(18,6))
```

```python
by_day_sentiment = df.groupby([pd.Grouper(key='date', freq=None), 'sentiment']) \
.size().unstack('sentiment')
```

```python
by_day_sentiment.plot()
#plt.title('Number of tweets per month and corrsponding sentiment')
#plt.title('Number of tweets per month and corrsponding sentiment from  to ')
```

```python
df['date'] = df['date'].dt.date
sentiment = df.groupby(['date', 'sentiment']).sentiment.count().unstack()
sentiment.plot(kind='bar')
plt.title('Announcements of Pfizer vaccine approval in 2020 ')
#plt.xlabel("")
#plt.ylabel("")
```

```python
plot_size = plt.rcParams["figure.figsize"] 
print(plot_size[0]) 
print(plot_size[1])

plot_size[0] = 8
plot_size[1] = 6
plt.rcParams["figure.figsize"] = plot_size 
df.sentiment.value_counts().plot(kind='pie', autopct='%1.0f%%')
```
